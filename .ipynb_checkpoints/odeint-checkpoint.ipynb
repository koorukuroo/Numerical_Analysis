```python
from scipy.integrate import odeint
```

# Example 1

http://bulldog2.redlands.edu/facultyfolder/deweerd/tutorials/Tutorial-ODEs.pdf
***

$$\frac{d^2y}{dt^2} = ay+b\frac{dy}{dt}$$

It can be rewritten as the following two first-order differential equations,
$$\frac{dy}{dt}=y' \quad\text{and}\quad \frac{dy'}{dt} = ay+by'$$

$y$ will be the first element $y[0]$ and $y'$ will be the second element $y[1]$.

That is,
$$\frac{dy[0]}{dt} = y[1] \quad\text{and}\quad \frac{dy[2]}{dt}=ay[0]+by[1]$$

```python
def deriv(y, t):
    a = -2.0
    b = -0.1
    return array([ y[1], a*y[0]+b*y[1] ])
```

```python
time = linspace(0, 10, 1000)
yinit = array([0.0005, 0.2])
y = odeint(deriv, yinit, time)
```

```python
plt.plot(time, y[:, 0])
plt.xlabel('t')
plt.ylabel('y')
```

# Example2

$$\frac{dy_1}{dt} = v_0 - k_1 y_1$$
$$\frac{dy_2}{dt} = k_1 y_1 - k_2 y_2$$

```python
v0 = 10
k1 = 0.5
k2 = 0.35
```

```python
def myModel(y, t):
    dy0 = v0 - k1*y[0]
    dy1 = k1*y[0] - k2*y[1]
    return [dy0, dy1]
```

```python
time = np.linspace(0, 20, 100)
yinit = array([0, 0])
y = odeint(myModel, yinit, time)
```

```python
plt.plot(time, y[:, 0], time, y[:, 1])
plt.xlabel('t')
plt.ylabel('y')
```

# Example 3

$$\frac{dk}{dt} = sAk^{\alpha} - (1-L+\delta)k$$
$$\frac{dL}{dt} = L(1-L)$$

$$\begin{cases}0 < \alpha < 1\\\delta >0 \end{cases}$$
$$s, A : contants$$

```python
s = 2
A = 1
delta = 2
alpha = 0.5
```

```python
def myModel(y, t):
    dk = s*A*y[0]**alpha - (1-y[1]+delta)*y[0]
    dL = y[1]*(1-y[1])
    return [dk, dL]
```

```python
time = np.linspace(0, 5, 1000)
yinit = array([10, 10])
y = odeint(myModel, yinit, time)
```

```python
plt.plot(time, y[:, 0], time, y[:, 1])
plt.xlabel('t')
plt.ylabel('y')
```

```python
for i in range(10):
    s = 2
    A = 1
    delta = 1
    alpha = 0.1*i

    def myModel(y, t):
        dk = s*A*y[0]**alpha - (1-y[1]+delta)*y[0]
        dL = y[1]*(1-y[1])
        return [dk, dL]

    time = np.linspace(0, 5, 1000)
    yinit = array([100, 100])
    y = odeint(myModel, yinit, time)
    plt.plot(time, y[:, 0], time, y[:, 1])
    plt.xlabel('t')
    plt.ylabel('y')
```

```python
plt.figure(figsize=(15, 15))

for i in range(10):
    s = 2
    A = 1
    delta = 1
    alpha = 0.1*i

    def myModel(y, t):
        dk = s*A*y[0]**alpha - (1-y[1]+delta)*y[0]
        dL = y[1]*(1-y[1])
        return [dk, dL]

    time = np.linspace(0, 5, 1000)
    yinit = array([100, 100])
    y = odeint(myModel, yinit, time)
    plt.plot(time, y[:, 0], time, y[:, 1], label='alpha '+str(0.1*i))
    
plt.legend(loc='best')
plt.xlabel('t')
plt.ylabel('y')
```
